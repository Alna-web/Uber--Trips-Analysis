```python
df['START_DATE'] = pd.to_datetime(df['START_DATE'], errors= 'coerce')
```

```python
df['END_DATE'] = pd.to_datetime(df['END_DATE'], errors= 'coerce')
```

```python
df.info()
```

```python
df['date'] = pd.DatetimeIndex(df['START_DATE']).date
```

```python
df['hour'] = pd.DatetimeIndex(df['START_DATE']).hour
```

```python
df['month'] = pd.DatetimeIndex(df['START_DATE']).month_name()
```

```python
df['day'] = pd.DatetimeIndex(df['START_DATE']).day_name()
```

```python
df.head()
```

```python
def time_period(hour):
    if 5 <= hour <=11:
        return 'morning'
    elif 12 <= hour <=15:
        return 'afternoon'
    elif 16 <= hour <=19:
        return 'evening'
    else:
        return 'night'
    df['time'] = df['hour'].apply(time_period)
```

```python
df.to_csv("cleaned_UberDataset.csv",index=False)
```
